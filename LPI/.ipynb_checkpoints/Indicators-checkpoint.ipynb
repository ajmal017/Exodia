```python
def formatTime(tim):
    if int(tim) < 10:
        return '0'+str(tim)
    else:
        return str(tim)
```

```python
def genOpenHoursTimes():
    lis = []
    for hour in range(9,16):
        for minute in range(0,60):
            lis += [formatTime(hour)+':'+formatTime(minute)]
    return lis[30:]+['16:00']
```

```python
# https://youtu.be/UfJ1HpzLmdk

# Compare the slope of the shareprice to the slope of the OBV to see whether there 
# is high price change low volume, or low price change high volume
def OBV(TICKER):
    if os.path.exists(f'./stock/1d/{TICKER}.json'):
        with open(f'./stock/1d/{TICKER}.json','r') as JSON:
            data = json.load(JSON)
    else:
        print('No price data for that ticker.')
        return

    if os.path.exists(f'./analysis/OBV') != True:
        os.mkdir(f'./analysis/OBV')
    
    KEYS = [i for i in data.keys()]
    dataList = [data[KEY] for KEY in KEYS]
    timestampsList = genOpenHoursTimes()
    OBVprev = 0
    OBVvals = {}
    cc = 0
    
    for ind in range(len(dataList)):
        for ts in range(len(timestampsList)):
            try:
                dataList[ind][timestampsList[ts]]['open']
            except:
                dataList[ind][timestampsList[ts]] = dataList[ind][timestampsList[ts-1]]
    
    for ind in range(1,len(dataList)):
        for timestamp in timestampsList:
            try:
                if dataList[ind][timestamp]['open'] > dataList[ind-1][timestamp]['open']:
                    OBVnow = OBVprev + dataList[ind][timestamp]['volumeDelta']
                elif dataList[ind][timestamp]['open'] < dataList[ind-1][timestamp]['open']:
                    OBVnow = OBVprev - dataList[ind][timestamp]['volumeDelta']
                else:
                    OBVnow = OBVprev + 0
                                
                try:
                    OBVvals[KEYS[ind]][timestamp] = OBVnow
                except:
                    OBVvals[KEYS[ind]] = {}
                    OBVvals[KEYS[ind]][timestamp] = OBVnow
                    
                OBVprev = OBVnow
            except:
                cc += 1
        
    print(f'Missing values {cc}')
    with open(f'./analysis/OBV/{TICKER}.json','w') as JSON:
        json.dump(OBVvals,JSON)
        
    return OBVvals
```

```python
def RSI(priceChangeList):
    gainList = [np.abs(i) for i in priceChangeList[0:13] if i > 0]
    lostList = [np.abs(i) for i in priceChangeList[0:13] if i < 0]
    avGain = sum(gainList)/len(gainList)
    avLost = sum(lostList)/len(lostList)
    
    RSIstart = 100-(100/(1 + avGain/avLost))
    RSIvals = [RSIstart]
    for ind in range(1,len(priceChangeList)-13):
        if priceChangeList[ind] >= 0:
            currentGain = priceChangeList[ind]
            currentLost = 0
        elif priceChangeList[ind] < 0:
            currentGain = 0
            currentLost = priceChangeList[ind]
            
        RSInow = 100 - (100/(1 + (avGain*13 + currentGain)/(avLost*13 + currentLost)))
        RSIvals += [RSInow]
        
        gainList = [np.abs(i) for i in priceChangeList[ind:13+ind] if i > 0]
        lostList = [np.abs(i) for i in priceChangeList[ind:13+ind] if i < 0]
        
        try:
            avGain = sum(gainList)/len(gainList)
        except:
            avGain = 0
        try:
            avLost = sum(lostList)/len(lostList)
        except:
            avLost = 0
    
    return [None]*14 + RSIvals
```

```python
# https://youtu.be/UfJ1HpzLmdk

# Compare the slope of the shareprice to the slope of the OBV to see whether there 
# is high price change low volume, or low price change high volume
def BollingerBands(TICKER):
    if os.path.exists(f'./stock/1d/{TICKER}.json'):
        with open(f'./stock/1d/{TICKER}.json','r') as JSON:
            data = json.load(JSON)
    else:
        print('No price data for that ticker.')
        return
    
    if os.path.exists(f'./analysis/BollingerBands') != True:
        os.mkdir(f'./analysis/BollingerBands')
    
    KEYS = [i for i in data.keys()]
    dataList = [data[KEY] for KEY in KEYS]
    timestampsList = genOpenHoursTimes()
    BBvals = {}
    cc = 0
   
    for ind in range(len(dataList)):
        for ts in range(len(timestampsList)):
            try:
                dataList[ind][timestampsList[ts]]['open']
            except:
                dataList[ind][timestampsList[ts]] = dataList[ind][timestampsList[ts-1]]
    
    for ind in range(13,len(dataList)):
        for timestamp in timestampsList:
            try:
                priceList = [dataList[i][timestamp]['open'] for i in range(ind-13,ind+1)]
                MA = sum(priceList)/len(priceList)
                STD = np.std(priceList)

                BOLU = MA + 2*STD
                BOLD = MA - 2*STD
                formatted = {'BOLU':BOLU,'BOLD':BOLD}

                try:
                    BBvals[KEYS[ind]][timestamp] = formatted
                except:
                    BBvals[KEYS[ind]] = {}
                    BBvals[KEYS[ind]][timestamp] = formatted
            except:
                cc += 1
                
    print(f'Missing values {cc}')
    with open(f'./analysis/BollingerBands/{TICKER}.json','w') as JSON:
        json.dump(BBvals,JSON)
        
    return BBvals
```

```python
def SMA(TICKER,DAYS):
    if os.path.exists(f'./stock/1d/{TICKER}.json'):
        with open(f'./stock/1d/{TICKER}.json','r') as JSON:
            data = json.load(JSON)
    else:
        print('No price data for that ticker.')
        return
    
    if os.path.exists(f'./analysis/SMA{DAYS}') != True:
        os.mkdir(f'./analysis/SMA{DAYS}')
    
    KEYS = [i for i in data.keys()]
    dataList = [data[KEY] for KEY in KEYS]
    timestampsList = genOpenHoursTimes()
    SMAvals = {}
    cc = 0
    
    for ind in range(len(dataList)):
        for ts in range(len(timestampsList)):
            try:
                dataList[ind][timestampsList[ts]]['open']
            except:
                dataList[ind][timestampsList[ts]] = dataList[ind][timestampsList[ts-1]]

    for ind in range(DAYS,len(dataList)):
        for timestamp in timestampsList:
            try:
                priceList = [dataList[i][timestamp]['open'] for i in range(ind-DAYS,ind+1)]
                SMA = sum(priceList)/len(priceList)

                try:
                    SMAvals[KEYS[ind]][timestamp] = SMA
                except:
                    SMAvals[KEYS[ind]] = {}
                    SMAvals[KEYS[ind]][timestamp] = SMA
            except:
                cc += 1
                
    print(f'Missing values {cc}')
    with open(f'./analysis/SMA{DAYS}/{TICKER}.json','w') as JSON:
        json.dump(SMAvals,JSON)
        
    return SMAvals
```

```python
# https://youtu.be/UfJ1HpzLmdk

# Compare the slope of the shareprice to the slope of the OBV to see whether there 
# is high price change low volume, or low price change high volume
def EMA(TICKER,DAYS):
    if os.path.exists(f'./stock/1d/{TICKER}.json'):
        with open(f'./stock/1d/{TICKER}.json','r') as JSON:
            data = json.load(JSON)
    else:
        print('No price data for that ticker.')
        return
    
    if os.path.exists(f'./analysis/EMA{DAYS}') != True:
        os.mkdir(f'./analysis/EMA{DAYS}')
    
    KEYS = [i for i in data.keys()]
    dataList = [data[KEY] for KEY in KEYS]
    timestampsList = genOpenHoursTimes()
    EMAvals = {}
    cc = 0
    
    for ind in range(len(dataList)):
        for ts in range(len(timestampsList)):
            try:
                dataList[ind][timestampsList[ts]]['open']
            except:
                dataList[ind][timestampsList[ts]] = dataList[ind][timestampsList[ts-1]]
    
    EMAy = {}
    SMAy = {}
    s = 2
    smooth = s/(1+DAYS)
    for ind in range(DAYS,len(dataList)):
        for timestamp in timestampsList:
            
            if ind == DAYS:
                priceList = [dataList[i][timestamp]['open'] for i in range(ind-DAYS,ind+1)]
                SMA = sum(priceList)/len(priceList)
                EMAt = priceList[-1]*smooth + SMA*(1-smooth)
                EMAy[timestamp] = EMAt
            
            else:
                try:
                    EMAyv = EMAy[timestamp]
                    EMAt = dataList[ind][timestamp]['open']*smooth + EMAyv*(1-smooth)
                    EMAy[timestamp] = EMAt
                    
                    try:
                        EMAvals[KEYS[ind]][timestamp] = EMAt
                    except:
                        EMAvals[KEYS[ind]] = {}
                        EMAvals[KEYS[ind]][timestamp] = EMAt
                except:
                    cc += 1
                
    print(f'Missing values {cc}')
    with open(f'./analysis/EMA{DAYS}/{TICKER}.json','w') as JSON:
        json.dump(EMAvals,JSON)
        
    return EMAvals
```

```python
def ROC(TICKER,DAYS):
    if os.path.exists(f'./stock/1d/{TICKER}.json'):
        with open(f'./stock/1d/{TICKER}.json','r') as JSON:
            data = json.load(JSON)
    else:
        print('No price data for that ticker.')
        return
    
    if os.path.exists(f'./analysis/ROC{DAYS}') != True:
        os.mkdir(f'./analysis/ROC{DAYS}')
    
    KEYS = [i for i in data.keys()]
    dataList = [data[KEY] for KEY in KEYS]
    timestampsList = genOpenHoursTimes()
    ROCvals = {}
    cc = 0
    
    for ind in range(len(dataList)):
        for ts in range(len(timestampsList)):
            try:
                dataList[ind][timestampsList[ts]]['open']
            except:
                dataList[ind][timestampsList[ts]] = dataList[ind][timestampsList[ts-1]]

    for ind in range(DAYS,len(dataList)):
        for timestamp in timestampsList:
            try:
                oldOPEN = dataList[ind-DAYS][timestamp]['open']
                nowOPEN = dataList[ind][timestamp]['open']
                ROCv = ((nowOPEN - oldOPEN) / oldOPEN)*100

                try:
                    ROCvals[KEYS[ind]][timestamp] = ROCv
                except:
                    ROCvals[KEYS[ind]] = {}
                    ROCvals[KEYS[ind]][timestamp] = ROCv
            except:
                cc += 1
                
    print(f'Missing values {cc}')
    with open(f'./analysis/ROC{DAYS}/{TICKER}.json','w') as JSON:
        json.dump(ROCvals,JSON)
        
    return ROCvals
```

```python
def ROCSTD(TICKER,DAYS):
    if os.path.exists(f'./stock/1d/{TICKER}.json'):
        with open(f'./stock/1d/{TICKER}.json','r') as JSON:
            data = json.load(JSON)
    else:
        print('No price data for that ticker.')
        return
    
    if os.path.exists(f'./analysis/ROCSTD{DAYS}') != True:
        os.mkdir(f'./analysis/ROCSTD{DAYS}')
    
    KEYS = [i for i in data.keys()]
    dataList = [data[KEY] for KEY in KEYS]
    timestampsList = genOpenHoursTimes()
    ROCvals = {}
    cc = 0
    
    for ind in range(len(dataList)):
        for ts in range(len(timestampsList)):
            try:
                dataList[ind][timestampsList[ts]]['open']
            except:
                dataList[ind][timestampsList[ts]] = dataList[ind][timestampsList[ts-1]]

    for ind in range(DAYS,len(dataList)):
        for timestamp in timestampsList:
            try:
                prices = [dataList[i][timestamp]['open'] for i in range(ind-DAYS,ind+1)]
                ROClist = [((prices[i+1]-prices[i])/prices[i])*100 for i in range(len(prices)-1)]
                gainLoss = (prices[-1]/prices[-2])-1
#                 gainLoss = sum([prices[i+1]-prices[i] for i in range(DAYS-1)])/(len(prices)-1)
                STD = np.std(ROClist)
#                 Av = sum(ROClist)/len(ROClist)
                STDval = STD*gainLoss #(STD*dataList[ind][timestamp]['open'])
#                 STDval = (STD/dataList[ind][timestamp]['open'])*100
    
                try:
                    ROCvals[KEYS[ind]][timestamp] = STDval
                except:
                    ROCvals[KEYS[ind]] = {}
                    ROCvals[KEYS[ind]][timestamp] = STDval
            except:
                cc += 1
                
    print(f'Missing values {cc}')
    with open(f'./analysis/ROCSTD{DAYS}/{TICKER}.json','w') as JSON:
        json.dump(ROCvals,JSON)
        
    return ROCvals
```
