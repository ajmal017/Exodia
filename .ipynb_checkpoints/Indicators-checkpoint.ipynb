```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
############ VOLUME ############
```

```python
# https://youtu.be/UfJ1HpzLmdk

# Compare the slope of the shareprice to the slope of the OBV to see whether there 
# is high price change low volume, or low price change high volume
def OBV(closeList,volumeList):
    OBVprev = 0
    OBVvals = []
    
    for ind in range(1,len(closeList)):
        if closeList[ind] > closeList[ind-1]:
            OBVnow = OBVprev + volumeList[ind]
        elif closeList[ind] < closeList[ind-1]:
            OBVnow = OBVprev - volumeList[ind]
        else:
            OBVnow = OBVprev + 0
        OBVvals += [OBVnow]
        OBVprev = OBVnow
        
    return OBVvals
```

```python
length = 50
closeList = [random.randint(5,50) for i in range(length)]
volumeList = [random.randint(50,200) for i in range(length)]
```

```python
OBVvals = OBV(closeList,volumeList)
```

```python
plt.plot([i for i in range(len(OBVvals))],OBVvals,c='r')
plt.plot([i for i in range(len(closeList))],closeList,c='b')
plt.plot([i for i in range(len(volumeList))],volumeList,c='g')
```

```python
def RSI(priceChangeList):
    gainList = [np.abs(i) for i in priceChangeList[0:13] if i > 0]
    lostList = [np.abs(i) for i in priceChangeList[0:13] if i < 0]
    avGain = sum(gainList)/len(gainList)
    avLost = sum(lostList)/len(lostList)
    
    RSIstart = 100-(100/(1 + avGain/avLost))
    RSIvals = [RSIstart]
    for ind in range(1,len(priceChangeList)-13):
        if priceChangeList[ind] >= 0:
            currentGain = priceChangeList[ind]
            currentLost = 0
        elif priceChangeList[ind] < 0:
            currentGain = 0
            currentLost = priceChangeList[ind]
            
        RSInow = 100 - (100/(1 + (avGain*13 + currentGain)/(avLost*13 + currentLost)))
        RSIvals += [RSInow]
        
        gainList = [np.abs(i) for i in priceChangeList[ind:13+ind] if i > 0]
        lostList = [np.abs(i) for i in priceChangeList[ind:13+ind] if i < 0]
        
        try:
            avGain = sum(gainList)/len(gainList)
        except:
            avGain = 0
        try:
            avLost = sum(lostList)/len(lostList)
        except:
            avLost = 0
    
    return [0]*13 + RSIvals
```

```python
length = 50
priceList = [random.uniform(40.0,50.0) for i in range(length)]
priceChangeList = [((priceList[i]/priceList[i-1])-1)*100 for i in range(1,len(priceList))]
```

```python
RSIvals = RSI(priceChangeList)
```

```python
plt.plot([i for i in range(len(RSIvals))],RSIvals,c='r')
plt.plot([i for i in range(len(priceList))],priceList,c='b')
#plt.plot([i for i in range(len(priceChangeList))],priceChangeList,c='g')
```
