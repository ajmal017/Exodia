```python
import time
import csv

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

from bs4 import BeautifulSoup
```

```python
class GenerateTickerList(object):
    def __init__(self):
        self.placeholder = 'placeholder'
        self.Stocktwits_URL = f'https://api.stocktwits.com/api/2/streams/symbol/{self.placeholder}.json?filter=top'

    
    def genNYSE(self):
        nyseTickers = []
        with open('nyse.csv') as CSV:
            csvFile = csv.reader(CSV, delimiter=',')
            for row in csvFile:
                if '$' not in row[0]:
                    nyseTickers += [row[0]]
        return nyseTickers[1:]
    
    
    def genNASDAQ(self):
        nasdaqTickers = []
        with open('nasdaq.csv') as CSV:
            csvFile = csv.reader(CSV, delimiter=',')
            for row in csvFile:
                nasdaqTickers += [row[0]]
        return nasdaqTickers[1:]
    
    
    def getNYSEtickers(self):
        nyseTickers = self.genNYSE()
        agent = {"User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/83.0.4103.61 Chrome/83.0.4103.61 Safari/537.36'}
        allTime = time.time()
        allResponse200 = 0
        allResponse404 = 0
        
        chrome_options = Options()
        chrome_options.add_argument("--headless")
        
        for struct in range((len(nyseTickers)//100)+1):
            driver = webdriver.Chrome(ChromeDriverManager().install(),options=chrome_options)
            t = time.time()
            response200 = 0
            response404 = 0
            stocktwitsTickers = []
            for tick in nyseTickers[struct*100:(struct+1)*100]:
                URL = f'https://stocktwits.com/symbol/{tick}'
                driver.get(URL)
                HTML = str(BeautifulSoup(driver.page_source.encode("utf-8")))
                print(tick)
                if "Uh-oh, we can\'t find this one" not in HTML:
                    stocktwitsTickers += [tick]
                    response200 += 1
                    print('[200]')
                else:
                    response404 += 1
                    print('---------- [400]')
            allResponse200 += response200
            allResponse404 += response404
            print(f'\n{struct} done in {time.time()-t}s')
            print(f'Response [200]: {response200}')
            print(f'Response [404]: {response404}')
            
            with open('nyseTwits.txt','a') as TXT:
                TXT.write(str(stocktwitsTickers))
                TXT.close()
            driver.close()
        
        print(f'\n\nALL DONE IN {(time.time()-allTime)/60}min')
        return stocktwitsTickers
    
    
    def getNASDAQtickers(self):
        nasdaqTickers = self.genNASDAQ()
        agent = {"User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/83.0.4103.61 Chrome/83.0.4103.61 Safari/537.36'}
        allTime = time.time()
        allResponse200 = 0
        allResponse404 = 0
        
        chrome_options = Options()
        chrome_options.add_argument("--headless")
        
        for struct in range((len(nasdaqTickers)//100)+1):
            driver = webdriver.Chrome(ChromeDriverManager().install(),options=chrome_options)
            stocktwitsTickers = []
            t = time.time()
            response200 = 0
            response404 = 0
            for tick in nasdaqTickers[struct*100:(struct+1)*100]:
                URL = f'https://stocktwits.com/symbol/{tick}'
                driver.get(URL)
                HTML = str(BeautifulSoup(driver.page_source.encode("utf-8")))
                print(tick)
                if "Uh-oh, we can\'t find this one" not in HTML:
                    stocktwitsTickers += [tick]
                    response200 += 1
                    print('[200]')
                else:
                    response404 += 1
                    print('---------- [400]')
            allResponse200 += response200
            allResponse404 += response404
            print(f'\n{struct} done in {time.time()-t}s')
            print(f'Response [200]: {response200}')
            print(f'Response [404]: {response404}')
            
            with open('nasdaqTwits.txt','a') as TXT:
                TXT.write(str(stocktwitsTickers))
                TXT.close()
            driver.close()

        print(f'\n\nALL DONE IN {(time.time()-allTime)/60}min')
        return stocktwitsTickers

    
    def getStocktwitsTickers(self):
        nyseTickers = self.getNYSEtickers()
        nasdaqTickers = self.getNASDAQtickers()
        
        return nyseTickers,nasdaqTickers
```
